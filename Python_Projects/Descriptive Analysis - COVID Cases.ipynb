### 2. Preprocessing and Cleansing of data

```python

#1. Checking the data types, rectifying the date format.
print("Data types of the fields of Covid Deaths:\n",df_covid_deaths.dtypes)
print("\n") # Adding line to have clarity in the output.
print("Data types of the fields of Covid Vaccinations:\n",df_covid_vacc.dtypes)

#Creating a copy of dataframe.
df_covid_deaths_date_formtt = df_covid_deaths.copy()
df_covid_vacc_date_formtt = df_covid_vacc.copy()

#All of the data types are showing as expected except the date field in both the dataframes.
df_covid_deaths_date_formtt['date'] = pd.to_datetime(df_covid_deaths_date_formtt['date'],format='%m/%d/%Y')
df_covid_vacc_date_formtt['date'] = pd.to_datetime(df_covid_vacc_date_formtt['date'],format='%m/%d/%Y')
print("\n")  #Adding line to have clarity in the output.

#After the dates are formatted.
print("Data types of the fields of Covid Deaths after Date formatting:\n",df_covid_deaths_date_formtt.dtypes)
print("\n") # Adding line to have clarity in the output.
print("Data types of the fields of Covid Vaccinations  after Date formatting:\n",df_covid_vacc_date_formtt.dtypes)
```

```python
#2. There are null values in certain fields but since the data is given in the chronological order 
    #and the missing data might signify the number of vaccinations, deaths or cases not recorded. 
    #Secondly, and equally important reason, when joining these two dataframes depending on the requirement would pose challenge.

#Definitely removing the fields from covid_vaccinations for which I couldn't find any description, 
#therefore, doesn't make sense to utilise in the analysis.

#Again creating copy of the dataframe.
df_covid_vacc_cleaned = df_covid_vacc_date_formtt.copy()
df_covid_deaths_cleaned = df_covid_deaths_date_formtt.copy()

#Removing the records where continents are null in both the dataframe.
df_covid_vacc_cleaned = df_covid_vacc_cleaned.dropna(subset=['continent'])
df_covid_deaths_cleaned = df_covid_deaths_cleaned.dropna(subset=['continent'])

#Dropping tables from Covid_Vaccinations dataframe.
df_covid_vacc_cleaned = df_covid_vacc_cleaned.drop(['stringency_index','population_density','median_age','aged_65_older','aged_70_older',
                                                    'gdp_per_capita','extreme_poverty','cardiovasc_death_rate','diabetes_prevalence','female_smokers','male_smokers',
                                                    'handwashing_facilities','hospital_beds_per_thousand','life_expectancy','human_development_index','excess_mortality_cumulative_absolute',
                                                    'excess_mortality_cumulative','excess_mortality','excess_mortality_cumulative_per_million'],axis = 1)



print("Covid Vaccinations table info:",df_covid_vacc_cleaned.info())  #Printing out the table info.
print("\n") #Adding to have clarity in the output.
print("Covid Deaths table info:",df_covid_deaths_cleaned.info())   #Printing out the table info.
```

### 3. Creation of Data Visualisations
From this section onwards, there are a couple of visualisations to come up with some insights on the trends of the cases and deaths happened in the countries. We will also see effects of vaccinations on the cases and deaths. In addition to this we will see which continent was most affected.

#### 3.1 Total Death Count and Cases in Continent - Bar Chart
This graph is a depiction of the high level statistics of COVID related cases and deaths at the continent level. As the data is worth almost 3 years. There can be certain of confidence in the data because of the source and the volume of the data. I have applied filter on threshold on the number of cases and deaths to see which continents are coming with higher figures.

```python
#The total cases and deaths can be obtained by adding new_cases and new_deaths grouped by continent.
df_death_total_cases_deaths = df_covid_deaths_cleaned.groupby(['continent']).agg({'new_cases':'sum','new_deaths':'sum'}).reset_index()

#Renaming the columns
df_death_total_cases_deaths.columns = ['continent','total_cases','total_deaths']

print("Total number of cases and deaths:\n",df_death_total_cases_deaths)
print("\n") #Adding a line to bring clarity in the output.

#Invoking library used for scale formatting.
from matplotlib.ticker import ScalarFormatter

# Set the position of the bars on the x-axis
bar_width = 0.35
r1 = range(len(df_death_total_cases_deaths['continent']))
r2 = [x + bar_width for x in r1]

# Create a figure and a set of subplots
fig, ax1 = plt.subplots()

# Plot the total cases on the primary y-axis
ax1.bar(r1, df_death_total_cases_deaths['total_cases'], color='blue', width=bar_width, edgecolor='grey', label='Total Number of Cases')
ax1.set_xlabel('Continent', fontweight='bold')
ax1.set_ylabel('Total Number of Cases', color='blue', fontweight='bold')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticks([r + bar_width/2 for r in range(len(df_death_total_cases_deaths['continent']))])
ax1.set_xticklabels(df_death_total_cases_deaths['continent'], rotation=45, ha='right')

# Remove scientific notation
ax1.yaxis.set_major_formatter(ScalarFormatter())
ax1.get_yaxis().get_major_formatter().set_scientific(False)
ax1.get_yaxis().get_major_formatter().set_useOffset(False)

# Create a secondary y-axis
ax2 = ax1.twinx()
ax2.bar(r2, df_death_total_cases_deaths['total_deaths'], color='green', width=bar_width, edgecolor='grey', label='Total Number of Deaths')
ax2.set_ylabel('Total Number of Deaths', color='green', fontweight='bold')
ax2.tick_params(axis='y', labelcolor='green')

# Remove scientific notation
ax2.yaxis.set_major_formatter(ScalarFormatter())
ax2.get_yaxis().get_major_formatter().set_scientific(False)
ax2.get_yaxis().get_major_formatter().set_useOffset(False)

# Add title and legend
plt.title('Total Number of COVID-19 Cases and Deaths by Continent')
fig.tight_layout()

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(1.5,0.75))

# Adjust layout to prevent overlap
plt.subplots_adjust(bottom=0.3)

# Add caption
fig.text(0.5, 0.01, 'Figure 3.1.1: Comparison of COVID-19 Cases and Deaths by Continent', ha='center', fontsize=10)

# Display the chart
plt.show()

# Adding filters 
# Filter and plot total cases
threshold_cases = input("Enter the threshold cases:")
threshold_cases = int(threshold_cases)
print("Threshold values of cases entered:",threshold_cases)

filtered_df_cases = df_death_total_cases_deaths[df_death_total_cases_deaths['total_cases'] > threshold_cases]

fig, ax = plt.subplots()
filtered_df_cases.groupby('continent')['total_cases'].sum().sort_values(ascending=True).plot.barh(
    ax=ax, figsize=(8, 6), title=f"Total Number of COVID-19 Cases by Continent with Threshold {threshold_cases}", color='blue')
ax.set_xlabel('Total Number of Cases')
ax.xaxis.set_major_formatter(ScalarFormatter())
ax.get_xaxis().get_major_formatter().set_scientific(False)
ax.get_xaxis().get_major_formatter().set_useOffset(False)
plt.show()

# Filter and plot total deaths
threshold_deaths = input("Enter the threshold deaths:")
threshold_deaths = int(threshold_deaths)
print("Threshold values of cases entered:",threshold_deaths)

filtered_df_deaths = df_death_total_cases_deaths[df_death_total_cases_deaths['total_deaths'] > threshold_deaths]

fig, ax = plt.subplots()
filtered_df_deaths.groupby('continent')['total_deaths'].sum().sort_values(ascending=True).plot.barh(
    ax=ax, figsize=(8, 6), title=f"Total Number of COVID-19 Deaths by Continent with Threshold {threshold_deaths}", color='green')
ax.set_xlabel('Total Number of Deaths')
ax.xaxis.set_major_formatter(ScalarFormatter())
ax.get_xaxis().get_major_formatter().set_scientific(False)
ax.get_xaxis().get_major_formatter().set_useOffset(False)

# Add caption
fig.text(0.5, 0.01, 'Figure 3.1.2: Bar Charts with filter on Total Cases and Deaths', ha='center', fontsize=10)

plt.show()

#Note: Took help from Copilot MS Edge AI assistant in getting help on beautifying and adding captions to the graphs.
```

#### 3.2 In a continent which country has maximum suffered - Pie Chart
This graph depicts the high level statistics on which country is maximum suffered in terms of the number of COVID cases. Taking a filter on the South American continent, the data depicts the maximum number of cases were from Brazil. It was more than 50% of the total of South American COVID cases.

```python
#Determining the total deaths in a country.
df_total_deaths_countries = df_covid_deaths_cleaned.groupby(['continent','location']).agg({'new_cases':'sum'}).reset_index()

#Merging the continent level dataframe and country level dataframe.
df_comp_country_continent = pd.merge(df_total_deaths_countries,df_death_total_cases_deaths, on='continent')

#Renaming the columns.
df_comp_country_continent.columns = ['continent','country','country_level_total_cases','continent_level_total_cases','continent_level_total_deaths']

#Getting a copy to further perform the required operations.
df_percent_country_continent = df_comp_country_continent.copy()

#Calculating the Percentage of the country cases out of Continent level cases.
df_percent_country_continent['percentage_contribution_of_cases']=round((df_percent_country_continent['country_level_total_cases']/df_percent_country_continent['continent_level_total_cases'])*100,2)


#Creating Pie-chart
#Filtering on Continent 
continent = input("Enter the Continent:")
print("Entered Continent is:",continent)
filtered_on_continent = df_percent_country_continent[df_percent_country_continent['continent'] == continent]


# Create the pie chart with country names as labels
plt.figure(figsize=(8, 8))
wedges, texts = plt.pie(filtered_on_continent['percentage_contribution_of_cases'], labels=filtered_on_continent['country'], 
                        autopct=None, pctdistance=0.85)

# Add a legend
plt.legend(wedges, filtered_on_continent['country'] + ' - ' + filtered_on_continent['percentage_contribution_of_cases'].astype(str)
           + '%', title="Countries", loc="center left", bbox_to_anchor=(1.5, 0, 0.5, 1))

# Add a title and caption
plt.title(f'Percentage of Country Level Cases in Continent {continent}')
plt.figtext(0.5, 0.01, 'Figure 3.2: This pie chart shows the percentage contribution of cases by country.', ha='center', fontsize=10)

plt.show()

#Note: Took help from MS Copilot on adjusting the graph so that this will not look cluttered.
```

#### 3.3 Correlation between Vaccinations and Cases - HeatMap
In this graph, a correlation is attempted to draw between a pair of countries, by how much increase in cases in one country has an effect on the other country. I have taken a filter of Oceania countries and tried to see that correlation. Majority of them is towards the spectrum that there is no correlation but there are some pair, for example, Vanuatu and Cook Islands or Samoa and Cook Islands seems to have some correlation. Looking at the google map, they are all near by nations. There might be some people moved among those countries but there is no evidence of those movement of people in the dataset. But one is sure that there are correlations of these numbers.

```python
#The number of vaccinations and the cases are in two different dataframes. They need to be joined on country, continent and date.
df_cases_sel_cols = df_covid_deaths_cleaned[['iso_code','continent','location','date','new_cases']]
df_vacc_sel_cols = df_covid_vacc_cleaned[['iso_code','continent','location','date','people_vaccinated']]


#Merging the new dataframes.
df_merged = pd.merge(df_cases_sel_cols,df_vacc_sel_cols, on =['iso_code','continent','location','date'])

#Renaming the Dataframes.
df_merged.columns = ['iso_code','continent','country', 'date','cases','vaccinations']


# Filter by continent
continent = input("Enter the Continent:")
print("Entered Continent is:",continent)

#Filter by dates - Start and End Date
start_date = input("Enter Start Date in the format YYYY-MM-DD:")
start_date = pd.to_datetime(start_date, format='%Y-%m-%d')
print("Entered Start Date is:",start_date)

end_date = input("Enter End Date in the format YYYY-MM-DD:")
end_date = pd.to_datetime(end_date, format='%Y-%m-%d')
print("Entered End Date is:",end_date)

df_continent = df_merged[(df_merged['continent'] == continent) & (df_merged['date'] >= start_date) & (df_merged['date'] <= end_date)]

# Pivot the DataFrame
df_pivot = df_continent.pivot(index='date', columns='country', values='cases')

# Calculate the correlation matrix
corr_matrix = df_pivot.corr()

# Create the heatmap
plt.figure(figsize=(20, 14))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=.5, cbar_kws={'ticks': [i/10 for i in range(-10, 11)]})
plt.title(f'COVID Cases Correlation Heatmap for {continent} from {start_date.date()} to {end_date.date()}')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

#Adding caption to the graph.
caption_text = 'Figure 3.3: Heatmap shows the Correlation of COVID cases between Countries'
plt.figtext(0.5, -0.05, caption_text, ha='right', fontsize=15)

plt.show()

#Note: Took help with beautifying the graph and having multiple filters.
```

#### 3.4 Correlation between Vaccinations and Cases - Scatter Plot
In this graph, a relationship between number of people vaccinated and the COVID cases is being attempt to draw. I am using the same data frame from the previous section. Followed by applying aggregation function of sum on vaccination numbers and cases groupby on continent, country, month and year. The month and year are extracted from the date present in the dataframe. I have applied the logarithmic function on both the vaccinations as well as the cases as they are huge in numbers. In the example below, I have applied filter on the continent Asia with dates from May in 2021 to October end, the vaccinations doesn't seem to be effective in bringing down the COVID cases but may be the symptoms may be less intense. In the current dataset, there is no evidence though. With this graph, different continent, country and date combination filters can be applied to view the relationship between the vaccinations and cases.

```python
#Using the same dataframe as in Section 3.3 df_merged for the graph.


#Extracting the Month and Year to apply the sum at this level
df_merged['month'] = df_merged['date'].dt.month
df_merged['year'] = df_merged['date'].dt.year

#Summing up to the level of Country, month and year.
df_agg = df_merged.groupby(['continent','country','month','year']).agg({'cases':'sum','vaccinations':'sum'}).reset_index()





#Directly applying the filters
# Filter by continent
continent = input("Enter the Continent:")
print("Entered Continent is:",continent)

# Filter by country
country = input("Enter the Country:")
print("Entered Continent is:",country)

#Filter by dates - Start and End Date
start_date = input("Enter Start Date in the format YYYY-MM-DD:")
start_date = pd.to_datetime(start_date, format='%Y-%m-%d')
print("Entered Start Date is:",start_date)

end_date = input("Enter End Date in the format YYYY-MM-DD:")
end_date = pd.to_datetime(end_date, format='%Y-%m-%d')
print("Entered End Date is:",end_date)

#Invoking the library to use the LogFormatterExponent function.
from matplotlib.ticker import LogFormatterExponent

# Apply filters
if country:
    df_cont_country_date = df_agg[(df_agg['continent'] == continent) & 
                                  (df_agg['country'] == country) & 
                                  ((df_agg['year'] > start_date.year) | 
                                   ((df_agg['year'] == start_date.year) & (df_agg['month'] >= start_date.month))) &
                                  ((df_agg['year'] < end_date.year) | 
                                   ((df_agg['year'] == end_date.year) & (df_agg['month'] <= end_date.month)))]
else:
    df_cont_country_date = df_agg[(df_agg['continent'] == continent) & 
                                  ((df_agg['year'] > start_date.year) | 
                                   ((df_agg['year'] == start_date.year) & (df_agg['month'] >= start_date.month))) &
                                  ((df_agg['year'] < end_date.year) | 
                                   ((df_agg['year'] == end_date.year) & (df_agg['month'] <= end_date.month)))]

#Creation of Scatter-plot
plt.figure(figsize=(10, 6))
plt.scatter(df_cont_country_date['cases'], df_cont_country_date['vaccinations'], alpha=0.5)
plt.xscale('log')
plt.yscale('log')
plt.title(f'Scatter Plot of Cases vs Vaccinations by Continent {continent},' + (f' Country {country} ' if country else '') 
          + f'between {start_date.date()} and {end_date.date()}')
plt.xlabel('Log of Number of Cases')
plt.ylabel('Log of Number of Vaccinations')
plt.grid(True, which="both", ls="-")

# Format the axis labels using LogFormatterExponent
plt.gca().xaxis.set_major_formatter(LogFormatterExponent())
plt.gca().yaxis.set_major_formatter(LogFormatterExponent())

# Add caption
plt.figtext(0.5, -0.1, 'Figure 3.4: Scatter-Plot to showcase the relationship between Logarithmic Vaccinations and Logarithmic COVID cases',
            wrap=True, horizontalalignment='center', fontsize=10)

plt.show()

#Note: Took help with beautifying the graph and having multiple filters.
```

```python
#Using a different dataframe to have vaccinations and deaths.
#Selecting the columns from the main data frames.
df_deaths_sel_cols1 = df_covid_deaths_cleaned[['iso_code','continent','location','date','new_deaths']]
df_vacc_sel_cols1 = df_covid_vacc_cleaned[['iso_code','continent','location','date','people_vaccinated']]

#Merging the new dataframes.
df_merged1 = pd.merge(df_deaths_sel_cols1,df_vacc_sel_cols1, on =['iso_code','continent','location','date'])
#Renaming the Columns.
df_merged1.columns = ['iso_code','continent','country', 'date','deaths','vaccinations']


#Extracting the Month and Year, I am summing up by country, month, year
df_merged1['month'] = df_merged1['date'].dt.month
df_merged1['year'] = df_merged1['date'].dt.year

#Summing by Country, Month and Year.
df_agg1 = df_merged1.groupby(['continent','country','month','year']).agg({'deaths':'sum','vaccinations':'sum'}).reset_index()





#Directly applying the filters
# Filter by continent
continent = input("Enter the Continent:")
print("Entered Continent is:",continent)

# Filter by country
country = input("Enter the Country:")
print("Entered Continent is:",country)

#Filter by dates - Start and End Date
start_date = input("Enter Start Date in the format YYYY-MM-DD:")
start_date = pd.to_datetime(start_date, format='%Y-%m-%d')
print("Entered Start Date is:",start_date)

end_date = input("Enter End Date in the format YYYY-MM-DD:")
end_date = pd.to_datetime(end_date, format='%Y-%m-%d')
print("Entered End Date is:",end_date)

#Invoking the library to use the FuncFormatter function.
from matplotlib.ticker import LogFormatterExponent

# Apply filters
if country:
    df_cont_country_date1 = df_agg1[(df_agg1['continent'] == continent) & 
                                  (df_agg1['country'] == country) & 
                                  ((df_agg1['year'] > start_date.year) | 
                                   ((df_agg1['year'] == start_date.year) & (df_agg1['month'] >= start_date.month))) &
                                  ((df_agg1['year'] < end_date.year) | 
                                   ((df_agg1['year'] == end_date.year) & (df_agg1['month'] <= end_date.month)))]
else:
    df_cont_country_date1 = df_agg1[(df_agg1['continent'] == continent) & 
                                  ((df_agg1['year'] > start_date.year) | 
                                   ((df_agg1['year'] == start_date.year) & (df_agg1['month'] >= start_date.month))) &
                                  ((df_agg1['year'] < end_date.year) | 
                                   ((df_agg1['year'] == end_date.year) & (df_agg1['month'] <= end_date.month)))]

#Creation of Scatter-plot
plt.figure(figsize=(10, 6))
plt.scatter(df_cont_country_date1['deaths'], df_cont_country_date1['vaccinations'], alpha=0.5)
#plt.xscale('log')
plt.yscale('log')
plt.title(f'Scatter Plot of Deaths vs Vaccinations by Continent {continent},' + (f' Country {country} ' if country else '') 
          + f'between {start_date.date()} and {end_date.date()}')
plt.xlabel('Number of Deaths')
plt.ylabel('Log of Number of Vaccinations')
plt.grid(True, which="both", ls="-")

# Custom formatter to remove exponential notation
plt.gca().yaxis.set_major_formatter(LogFormatterExponent())

# Add caption
plt.figtext(0.5, -0.1, 'Figure 3.5: Scatter-Plot to showcase the relationship between Logarithmic Vaccinations and Deaths',
            wrap=True, horizontalalignment='center', fontsize=10)

plt.show()
#Note: Took help with beautifying the graph and having multiple filters.
```

#### 3.6 Percentage Population Infected Throughout Time - Line Chart
In this graph, the growth in percentage of the population infected with COVID with time is depicted. It is achieved by taking the running sum of the cases in the dataframe related to the covid deaths. Followed by the calculation of the percentage of population infected by dividing the running sum by the population of the country. This graph depicts the rate of growth with time. Different countries can be taken into filter and the rate of growth can be studied to see which countries has the highest growth because that will be the cause of worry and precautionary measures can be advised to the government of the country. For example, in the below example, while studying the rate of growth, India seems to have the highest growth. Maximum caution has to be taken by the government of India.

```python
#In this chart, running sum will be applied on the cases and then dividing it by the population of the country.
#Taking selected columns and creating a new data frame.
df_deaths_sel_col2 = df_covid_deaths_cleaned[['iso_code','continent','location','date','new_cases','population']]

#Renaming the Columns.
df_deaths_sel_col2.columns = ['iso_code','continent','country','date','new_cases','population']

# Sort the dataframe by the grouping columns and date.
df_deaths_sel_col2 = df_deaths_sel_col2.sort_values(by=['iso_code', 'continent', 'country', 'date'])

# Calculate the running sum of new_cases.
df_deaths_sel_col2['running_sum_cases'] = df_deaths_sel_col2.groupby(['iso_code', 'continent', 'country'])['new_cases'].cumsum()

#Calculating the percentage of population infected.
df_deaths_sel_col2['percentage_population_infected'] = round(df_deaths_sel_col2['running_sum_cases']*100/df_deaths_sel_col2['population'],2)

# Collecting multiple countries for filtering
countries_filter = []
while True:
    country = input("Enter a country to add to the filter (or press Enter to finish):")
    if country:
        countries_filter.append(country)
    else:
        break

print(f"Countries entered: {countries_filter}")

if countries_filter:
    filtered_df = df_deaths_sel_col2[df_deaths_sel_col2['country'].isin(countries_filter)]
else:
    filtered_df = df_deaths_sel_col2

# Plotting the line chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=filtered_df, x='date', y='percentage_population_infected', hue='country')

plt.xlabel('Date')
plt.ylabel('Percentage of Population Infected')
plt.title('Percentage of Population Infected Over Time')
plt.grid(True)

# Adding the caption
plt.figtext(0.5, -0.1, f'Figure 3.6: Percentage of Population Infected with Time in {countries_filter}', wrap=True, horizontalalignment='center', fontsize=10)

plt.show()
```
